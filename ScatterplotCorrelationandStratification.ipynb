```r
#With two or more variables plots such as scatterplots and parameters like correlation coefficient can be useful
```

```r
library(UsingR)
data('father.son')
x <- father.son$fheight
y <- father.son$sheight
plot(x,y,xlab='Fathers height', ylab='Sons Height', main=paste('correlation =',signif(cor(x,y),2)))
```

```r
#Here we can view a trend, the taller the father, the taller the son
```

```r
#Let's stratify the father height in relation to the son
#What this means is since the father is taller than avg by a given number
#is the son taller than average by the same standard deviations?
groups <- split(y,round(x))
boxplot(groups)
```

```r
mean(y[round(x) == 72])
```

```r
#The average height of sons with fathers that are 72 inches is 70.7
```
