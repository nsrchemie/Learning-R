```r
#Revisiting falling object example
#Recall that the trajectory of a falling object is
#d = startingheight + startingvelocity - 0.5 * 9.8*time^2
#To simulate n observations we add measurement error
g <- 9.8 # gravity
n <- 25 #observations
tt <- seq(0,3.4,len=n) #Time, collected in numerous observations
f <- 56.67 - 0.5*g*tt^2
y <- f + rnorm(n,sd=1)
```

```r
plot(tt,y,ylab='distance in meters',xlab='time in seconds')
lines(tt,f,col=1)
```

```r
#here we are going to utilize the Least Squares equation
#this equation allows us to find values that minimize distance
#of the fitted model to the data
#We find the minimum, (the values known as the least squares estimates)
#evaluating the lse at the estimates gives us the residual sum of squares
#These are all random variables
```

```r
#Let's find the LSE for the falling object
#Also the standard errors and p-values
tt2 <- tt^2
fit <- lm(y~tt+tt2)
summary(fit)$coef
```
